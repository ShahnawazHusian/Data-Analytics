```python
#libraries to use
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#load dataset
df=pd.read_csv("C:/dataanalyst datasets/IPL Dataset.csv")
```

```python
#for first 5 rows
df.head()
```

```python
#for adding annotation on the bars
def addgraph(x,y):
    bbox_dict = {'facecolor': 'orange', 'alpha': 0.3}
    for i in range(len(x)):
        plt.text(i,y[i],y[i],fontsize = 12,ha = 'center',va = 'bottom',bbox = bbox_dict)
```

```python
#calculates the total runs scored by each bowler and displays the top 10 bowlers with the most runs
bo = df.groupby('bowler')
ru = bo['run'].sum().head(10).sort_values(ascending=False)
bowler_run = pd.DataFrame(ru)
bowler_run.reset_index('bowler',inplace=True)
```

```python
#plot the graph
plt.figure(figsize=(15,5))
plt.title('Bowler Who Conceded Most Runs',size=25,color='black')
sns.barplot(x='bowler',y='run',data=bowler_run)
plt.xlabel("Bowler's Name",size=20,color='darkblue')
plt.ylabel("Given Runs",size=20,color='darkred')
a = bowler_run['bowler']
b = bowler_run['run']
addgraph(a,b)
plt.show()
```

```python
bo1=df.groupby('bowler')
ru=bo1['run'].sum().head(10).sort_values(ascending=True)
bowler_run1=pd.DataFrame(ru)
bowler_run1.reset_index('bowler',inplace=True)
bowler_run1
a1=bowler_run1['bowler']
b1=bowler_run1['run']
```

```python
plt.figure(figsize=(15,5))
plt.title('Bowler Who Conceded Less Runs',size=25,color='red')
sns.barplot(x='bowler',y='run',data=bowler_run1)
plt.xlabel("Bowler's Name",size=20,color='green')
plt.ylabel("Given Runs",size=20,color='blue')
addgraph(a1,b1)
plt.show()
```

```python
se=df.groupby('season')
re=se['run'].sum().head(10).sort_values(ascending=False)
ses=pd.DataFrame(re)
ses.reset_index('season',inplace=True)
s1=ses['run']
s2=ses['season']
ses
```

```python
def addgraph2(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i],ha='center',va='bottom')
```

```python
plt.figure(figsize=(15,5))
plt.title('Total Runs in Season',size=25,color='orange')
sns.barplot(x='season',y='run',data=ses)
plt.xlabel("Season",size=20,color='purple')
plt.ylabel("Total Runs",size=20,color='brown')
addgraph2(s2,s1)
plt.show()
```

```python
bt=df.groupby('striker')
rt=bt['run'].sum().head(10).sort_values(ascending=False)
batter_run=pd.DataFrame(rt)
batter_run.reset_index('striker',inplace=True)
bt1=batter_run['striker']
bt2=batter_run['run']
```

```python
plt.figure(figsize=(15,5))
plt.title('Max Score By Batsman ',size=25,color='orange')
sns.barplot(x='striker',y='run',data=batter_run)
plt.xlabel("Batsman Name",size=20,color='purple')
plt.ylabel("Total Runs",size=20,color='brown')
addgraph2(bt1,bt2)
plt.show()
```

```python
rb = df.groupby('striker')
rb2 = rb['runs_off_bat'].sum().head(10).sort_values(ascending=False)
runoff = pd.DataFrame(rb2)
runoff.reset_index('striker',inplace=True)
runoff
rb1 = runoff['striker']
rb2 = runoff['runs_off_bat']
```

```python
plt.figure(figsize=(15,5))
plt.title('Max Runs By Bat',size=25,color='black')
sns.barplot(x='striker',y='runs_off_bat',data=runoff)
plt.xlabel("Batsman Name",size=20,color='purple')
plt.ylabel("Total Runs",size=20,color='brown')
addgraph2(rb1,rb2)
plt.show()
```

```python
ov=df.groupby('striker')
ov2=ov['over'].sum().head(10).sort_values(ascending=False)
overs=pd.DataFrame(ov2)
overs.reset_index('striker',inplace=True)
overs
o1=overs['striker']
o2=overs['over']
```

```python
plt.figure(figsize=(15,5))
plt.title('Max Over Played By Batsman ',size=25,color='black')
sns.barplot(x='striker',y='over',data=overs)
plt.xlabel("Batsman Name",size=20,color='purple')
plt.ylabel("Total Overs",size=20,color='brown')
addgraph2(o1,o2)
plt.show()
```

```python
w_df=df[df['wicket_type'].isin(['bowled'])]
wi=w_df.groupby('bowler')
wi3=wi['wicket_type'].value_counts().sort_values(ascending=False).head(10)
wicket=pd.DataFrame(wi3)
wicket.reset_index('bowler',inplace=True)

we=wicket['bowler']
we2=wicket['wicket_type']
```

```python
plt.figure(figsize=(15,5))
plt.title('Most Bowled By Bowler',size=25,color='black')
sns.barplot(x='bowler',y='wicket_type',data=wicket)
plt.xlabel("Bowler's Name",size=20,color='purple')
plt.ylabel("Total Bowled",size=20,color='brown')
addgraph2(we,we2)
plt.show()
```

```python
team = df.groupby('batting_team')
team2 = team['run'].sum().head(10).sort_values(ascending=False)
bteam = pd.DataFrame(team2)
bteam.reset_index('batting_team',inplace=True)
bt1 = bteam['batting_team']
bt2 = bteam['run']
```

```python
plt.figure(figsize=(15,5))
plt.title('Most Runs By Team In All Season',size=25,color='black')
plt.xticks(rotation=30,size=15)
sns.barplot(x='batting_team',y='run',data=bteam)
plt.xlabel("Teams Name",size=20,color='purple')
plt.ylabel("Total Runs",size=20,color='brown')
addgraph2(bt1,bt2)
plt.show()
```
