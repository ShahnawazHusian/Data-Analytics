```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("C:/dataanalyst datasets/crime dataset/crimeDatasets/10_Property_stolen_and_recovered.csv")
```

```python
df.head(3)
```

```python
def addgraph(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i],ha='center',va='bottom')
```

```python
a_n=df.groupby('Area_Name')
st=a_n['Cases_Property_Stolen'].sum().sort_values(ascending=False).head(10)
st2=pd.DataFrame(st)
st2.reset_index('Area_Name',inplace=True)
```

```python
plt.figure(figsize=(18,5))
plt.title('Cities with the most property theft cases',size=25,color='Red')
sns.barplot(x='Area_Name',y='Cases_Property_Stolen',data=st2)
plt.xlabel('City Name',size=20,color='brown')
plt.ylabel('Total Cases',size=20,color='purple')
addgraph(st2['Area_Name'],st2['Cases_Property_Stolen'])
plt.show()
```

```python
a_n=df.groupby('Area_Name')
re=a_n['Cases_Property_Recovered'].sum().sort_values(ascending=False).head(10)
re2=pd.DataFrame(re)
re2.reset_index('Area_Name',inplace=True)
```

```python
plt.figure(figsize=(18,5))
plt.title('Cities with the most solved property theft cases',size=25,color='orange')
sns.barplot(x='Area_Name',y='Cases_Property_Recovered',data=re2)
plt.xlabel('City Name',size=20,color='brown')
plt.ylabel('Solved Cases',size=20,color='purple')
addgraph(re2['Area_Name'],re2['Cases_Property_Recovered'])
plt.show()
```

```python
a_n=df.groupby('Area_Name')
stm=a_n['Cases_Property_Stolen'].sum().sort_values(ascending=True).head(10)
stm2=pd.DataFrame(stm)
stm2.reset_index('Area_Name',inplace=True)
```

```python
plt.figure(figsize=(18,5))
plt.title('Cities with the Lowest Property Theft Rates',size=25,color='Red')
sns.barplot(x='Area_Name',y='Cases_Property_Stolen',data=stm2)
plt.xlabel('City Name',size=20,color='brown')
plt.ylabel('Total Cases',size=20,color='purple')
addgraph(stm2['Area_Name'],stm2['Cases_Property_Stolen'])
plt.show()
```

```python
a_n=df.groupby('Area_Name')
rem=a_n['Cases_Property_Recovered'].sum().sort_values(ascending=True).head(10)
rem2=pd.DataFrame(rem)
rem2.reset_index('Area_Name',inplace=True)
```

```python
plt.figure(figsize=(18,5))
plt.title('Cities with the Highest Clearance Rates for Property Theft',size=25,color='orange')
sns.barplot(x='Area_Name',y='Cases_Property_Recovered',data=rem2)
plt.xlabel('City Name',size=20,color='brown')
plt.ylabel('Solved Cases',size=20,color='purple')
plt.xticks(rotation=20,size=12)
addgraph(rem2['Area_Name'],rem2['Cases_Property_Recovered'])
plt.show()
```

```python

ye=df.groupby('Year')
sy=ye['Cases_Property_Stolen'].sum().sort_values(ascending=False).head(10)
sy2=pd.DataFrame(sy)
sy2.reset_index('Year',inplace=True)
```

```python
plt.figure(figsize=(18,5))
plt.title('National Property Theft Trends',size=25,color='orange')
sns.barplot(x='Year',y='Cases_Property_Stolen',data=sy2)
plt.xlabel('Years',size=20,color='brown')
plt.ylabel('Total Cases',size=20,color='purple')
addgraph(sy2['Year'],sy2['Cases_Property_Stolen'])
plt.show()
```

```python
s_n=df.groupby('Sub_Group_Name')
sn=s_n['Value_of_Property_Stolen'].sum().sort_values(ascending=False).head(10)
stn2=pd.DataFrame(sn)
stn2.drop(['7. Total Property Stolen & Recovered'],axis=0,inplace=True)
stn2.reset_index('Sub_Group_Name',inplace=True)
```

```python
plt.figure(figsize=(18,5))
plt.title('Money Loss by Crime Type',size=25,color='Red')
sns.barplot(x='Sub_Group_Name',y='Value_of_Property_Stolen',data=stn2)
plt.xlabel('Crime Type',size=20,color='brown')
plt.ylabel('Total Money',size=20,color='purple')
addgraph(stn2['Sub_Group_Name'],stn2['Value_of_Property_Stolen'])
plt.show()
```

```python
s_n=df.groupby('Sub_Group_Name')
snr=s_n['Value_of_Property_Recovered'].sum().sort_values(ascending=False).head(10)
str2=pd.DataFrame(snr)
str2.drop(['7. Total Property Stolen & Recovered'],axis=0,inplace=True)
str2.reset_index('Sub_Group_Name',inplace=True)
```

```python
plt.figure(figsize=(18,5))
plt.title('Money Recovered By Crime Type',size=25,color='Red')
sns.barplot(x='Sub_Group_Name',y='Value_of_Property_Recovered',data=str2)
plt.xlabel('Crime Type',size=20,color='brown')
plt.ylabel('Recovered Money',size=20,color='purple')
addgraph(str2['Sub_Group_Name'],str2['Value_of_Property_Recovered'])
plt.show()
```
